# we will learn how to do embedding for a csv file

```python
import os
from getpass import getpass
from langchain_community.embeddings import HuggingFaceEmbeddings
```

```python
HUGGINGFACEHUB_API_TOKEN = getpass()
```

```python
os.environ["HUGGINGFACEHUB_API_TOKEN"] = HUGGINGFACEHUB_API_TOKEN
```

```python
# https://huggingface.co/sentence-transformers

# model_name = "sentence-transformers/all-mpnet-base-v2"
# model_kwargs = {'device': 'cpu'}
# encode_kwargs = {'normalize_embeddings': False}

embeddings = HuggingFaceEmbeddings()
```

```python
from langchain.document_loaders import CSVLoader
```

```python
loader = CSVLoader('iris.csv')
```

```python
data = loader.load()
```

```python
# we will just return string
data[0].page_content
```

```python
# get all string
[text.page_content for text in data]
```

```python
embedded_doc = embeddings.embed_documents([text.page_content for text in data])
```

```python
# list of vectors
embedded_doc
```

```python
len(embedded_doc)
# it is the total number of rows in our csv file
```
